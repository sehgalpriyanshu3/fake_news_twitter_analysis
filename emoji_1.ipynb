```python
import csv
df=pd.read_csv("twitter.csv",delimiter=",",encoding = "ISO-8859-1")
df = df.sample(frac=1).reset_index(drop=True)
dg=df.copy()
testing=df[0:24997]

testing.to_csv("testing2", sep='\t', encoding='utf-8')
training=df[24997:-1]
training.to_csv("training2", sep='\t', encoding='utf-8')
print(df)
```

```python
del df["ItemID"]
```

```python
d=df["SentimentText"]
data=d.values[0:40000]
data.shape
print(data)
```

```python
word_data=[]

for i in range(len(data)):
    word_data.append(data[i])
print(len(word_data))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create the transform
vectorizer = TfidfVectorizer()
# tokenize and build vocab
vectorizer.fit(word_data)
# summarize
vocab = vectorizer.vocabulary_
print(vectorizer.idf_)
# encode document
vector = vectorizer.transform(word_data)
# summarize encoded vector
print(vector.shape)
x_train=vector.toarray()


```

```python
y=df["Sentiment"].values[0:40000]
```

```python
from sklearn import model_selection
x_train1, x_test, y_train, y_test = model_selection.train_test_split(x_train, y, test_size = 0.25, random_state = 0)


```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(x_train1, y_train)
y_pred = clf.predict(x_test)

train_score = clf.score(x_train1, y_train)
test_score = clf.score(x_test, y_test)

print(test_score)

```
