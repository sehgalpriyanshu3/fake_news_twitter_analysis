```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
from sklearn.utils import shuffle


```

```python
#READING THE CSV FILE
tw1=pd.read_csv('predicted_sentiment.csv',delimiter=',')
#SHUFFLING THE DATA
tw1 = tw1.sample(frac=1).reset_index(drop=True)
twitter=tw1.copy()
twitter
```

```python
#READING THE CSV FILE
dg=pd.read_csv('train.tsv',delimiter='	')
#SHUFFLING THE DATA
dg = dg.sample(frac=1).reset_index(drop=True)
df=dg.copy()

df.columns=["ID","label","statement","subject","speaker","speakers_job_title","state_info","job_affiliation","barely_true","false","half_true","mostly_true","pants_on_fire","context"]
df
```

```python

unique=set(df["label"])
print(unique)
del df["ID"]

```

```python
df.speakers_job_title.replace(np.NaN, 'unknown', inplace=True)
df.state_info.replace(np.NaN, 'unknown', inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
def replace (s):
    if s=="mostly-true":
        return 0
    elif s=="barely-true":
        return 1
    elif s=="half-true":
        return 2
    elif s=="true":
        return 3
    elif s=="false":
        return 4
    elif s=="pants-fire":
        return 5
   
    
df["output"]=df.label.apply(replace)
output=df["output"]
unique=set(df["output"])
print(unique)
```

CREATING A TOKENIZED FORM OF DATA TO WORK UPON

```python
d=df["statement"]
statement=d.values
statement.shape
subject=df["subject"].values


```

```python
word_data=[]

for i in range(len(statement)):
    word_data.append(statement[i])
print(len(word_data))
```

STATEMENT TOKENIZATION

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# create the transform
vectorizer = TfidfVectorizer()
# tokenize and build vocab
vectorizer.fit(word_data)
# summarize
vocab = vectorizer.vocabulary_
print(vectorizer.idf_)
# encode document
vector = vectorizer.transform(word_data)
# summarize encoded vector
print(vector.shape)
x=vector.toarray()

print(vocab)



```

SUBJECT TOKENIZATION

```python
tweet=twitter["text"]
```

```python
tweets=[]

for i in range(len(tweet)):
    tweets.append(tweet[i])
print(len(tweets))
```

```python
vectorizer = TfidfVectorizer(vocabulary=vocab)
vectorizer.fit(tweets)
vector5 = vectorizer.transform(tweets)
# summarize encoded vector
print(vector5.shape)
x5=vector5.toarray()


```
